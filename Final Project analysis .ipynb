```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
wake_df = pd.read_csv("wake.csv", sep=",", dtype=str)
wake_df.head()
```

```python
cont_group = wake_df[wake_df['treatment'].str.contains('cont')]
print(len(cont_group))
```

```python
mel_group = wake_df[wake_df['treatment'].str.contains('mel')]
print(len(mel_group))
```

```python
cont = wake_df.iloc[0:96, [6]]
print(cont)
```

```python
mel = wake_df.iloc[96: , [6]]
print(mel)
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,5))

ax[0].plot(np.mean(cont, axis =1))
ax[1].plot(np.mean(mel, axis = 1))

ax[0].legend(['control'], loc = 'upper right')
ax[1].legend(['treatment'], loc = 'upper right')
```
